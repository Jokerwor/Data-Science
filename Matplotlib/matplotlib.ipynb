```python
x = np.random.rand(50)
y = np.random.rand(50)
plt.scatter(x,y)
```

```python
x
```

```python
y
```

```python
colours = np.random.rand(50)
sizes = 1000* np.random.rand(50)
plt.scatter(x, y,c=colours, s= sizes,alpha=0.5,cmap='viridis') 
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Data')
plt.show()
```

```python
x = ['a', 'b', 'c' ,'d', 'e']
y = np.random.rand(5)
plt.bar(x,y)
plt.Figure(figsize=(5, 5))
plt.grid()
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Data')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
colours = np.random.rand(50)
sizes = 1000* np.random.rand(50)
plt.scatter(x, y,c='r', s= sizes,alpha=0.5,cmap='viridis',marker='v') 
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Data')
plt.show()
```

```python
x = [48,75,64,76,81,39,10]
y = [19,32,23,46,85,76,57]
plt.plot(x,y)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Data')
plt.show()
```

```python
data = np.random.randn(20)
plt.hist(data)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Data')
plt.show()
```

```python
x = [14,65,36,87,48,69,10]
y = [72,63,43,50,76,87,98]
colour = ['red', 'blue', 'green', 'yellow', 'purple', 'orange', 'pink']
plt.scatter(x,y,c = colour)
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Random Data')
plt.show()
```

```python
x = np.random.rand(50)
y = np.random.rand(50)
z = np.random.rand(50)
fig = plt.figure()
ax = fig.add_subplot(projection = '3d')
ax.scatter(x,y,z)
plt.show()
```
